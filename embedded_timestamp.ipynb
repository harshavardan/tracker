```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("./video1.avi")
```

```python
px0 = []
px1 = []
px2 = []

while True:
    ret, frame = cap.read()
    if ret == 0:
        break
        
    cv2.imshow("frame", frame[:4, :4])
    px0.append(frame[0, 0])
    px1.append(frame[1, 0])
    px2.append(frame[2, 0])
    
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```
